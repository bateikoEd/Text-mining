**get_count_texts_in_period** 
<br> This function return count of articles in each period (type=np.array)

```python
def get_count_texts_in_period(result_tabes, dates_of_preiod):
    counts = np.array([])
    for i in range(len(dates_of_preiod)):
        df_temp = result_tabes[(dates_of_preiod['start_period'][i] <= result_tabes['date']) &
                               (result_tabes['date'] <= dates_of_preiod['end_period'][i])]['file_code']
        count = len(set(df_temp))
        counts = np.append(counts, values=count )
        print(count)
    return counts
```

```python
counts_each_period = get_count_texts_in_period(new_csv1, dates)
counts_each_period
```
